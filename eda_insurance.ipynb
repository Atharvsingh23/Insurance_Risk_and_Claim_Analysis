Insurance Risk & Claim EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dataset

```python
file_path = "insurance_policies_data.xlsx"   # filename in quotes
df = pd.read_excel(file_path, sheet_name=0)  # use the variable here

print(df.head())
```

Baic Insights


```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nInfo:")
print(df.info())
print("\nSummary Stats:")
print(df.describe(include="all"))
```

Missing Values


```python
plt.figure(figsize=(8,5))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

Gender Analysis

```python
plt.figure(figsize=(6,4))
df["Gender"].value_counts().plot(kind="pie", autopct="%.1f%%", colors=["lightblue","pink"], startangle=90)
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Gender", y="Claim Amount", data=df, estimator=sum, ci=None, palette="Set2")
plt.title("Total Claim Amount by Gender")
plt.ylabel("Total Revenue (Claim Amount)")
plt.show()
```

Education Analysis

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Education", y="Claim Amount", data=df, estimator=sum, ci=None, palette="muted")
plt.title("Total Claim Amount by Education Level")
plt.ylabel("Total Revenue")
plt.xticks(rotation=30)
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x="Education", y="Claim Amount", data=df, palette="pastel")
plt.title("Revenue Distribution by Education")
plt.xticks(rotation=30)
plt.show()
```

Coverage Zone

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Coverage Zone", y="Claim Amount", data=df, estimator=sum, ci=None, palette="coolwarm")
plt.title("Total Claim Amount by Coverage Zone")
plt.xticks(rotation=30)
plt.show()
```

Car Use

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Car Use", y="Claim Amount", data=df, estimator=sum, ci=None, palette="husl")
plt.title("Total Claim Amount by Car Use")
plt.show()
```

 Car Year (Trend over time)

```python
car_year_revenue = df.groupby("Car Year")["Claim Amount"].sum().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(x="Car Year", y="Claim Amount", data=car_year_revenue, marker="o")
plt.title("Total Claim Amount by Car Year")
plt.ylabel("Total Revenue")
plt.show()
```

Household Income

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Household Income"], bins=30, kde=True, color="teal")
plt.title("Distribution of Household Income")
plt.show()
```

Income vs Revenue

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Household Income", y="Claim Amount", hue="Gender", data=df, alpha=0.5)
plt.title("Household Income vs Claim Amount")
plt.show()
```

Correlation Heatmap (numeric cols)

```python
plt.figure(figsize=(8,6))
corr = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, cmap="Blues", fmt=".2f")
plt.title("Correlation Heatmap (Numeric Columns)")
plt.show()
```

Age Analysis (from BirthDate)

```python
df["Age"] = (pd.Timestamp("today").year - df["BirthDate"].dt.year)
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=30, kde=True, color="orange")
plt.title("Age Distribution of Policy Holders")
plt.show()
```

```python
age_revenue = df.groupby(pd.cut(df["Age"], bins=[18,30,40,50,60,70,80]))["Claim Amount"].sum()
age_revenue.plot(kind="bar", figsize=(8,5), color="skyblue")
plt.title("Total Claim Amount by Age Group")
plt.ylabel("Total Revenue")
plt.show()
```

Kids Driving Impact

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Kids Driving", y="Claim Amount", data=df, estimator=sum, ci=None, palette="Set3")
plt.title("Total Claim Amount by Kids Driving")
plt.show()
```

Pivot Summary

```python
pivot_summary = df.pivot_table(values="Claim Amount", index="Education", columns="Gender", aggfunc="sum")
print("\n=== Revenue Summary by Education  Gender ===")
print(pivot_summary)
```
